```python
import pandas as pd
import numpy as np 
from pandas_datareader.data import DataReader
from datetime import datetime
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
symbol='NKE'    
end=datetime.now()
start='2015-01-01'
dataset = DataReader(symbol, 'yahoo', start, end)
ChartData=dataset.reset_index()
ChartData.insert(1,'Days',ChartData.index+1)
# ChartData=ChartData.filter(['Days','Close'])
# # print('Total dataset Length',len(ChartData))
# print(ChartData)
# # ChartData=ChartData.values

import warnings
warnings.filterwarnings("ignore")

ChartData['MovingAvg']=ChartData['Close'].rolling(window=7).mean()
ChartData = ChartData.iloc[6:]
print(ChartData.head(10))



training_data_len = int(np.ceil( len(ChartData) * .70 ))
# print(nparry)
X_train = ChartData.iloc[0:training_data_len, 1:2].values
# X_train=X_train.values
y_train = ChartData.iloc[0:training_data_len, 5:6].values

X_test = ChartData.iloc[training_data_len:, 1:2].values
# X_train=X_train.values
y_test = ChartData.iloc[training_data_len:, 5:6].values

# TrainData=pd.read_csv('TSLA_train3.csv')
# TestData=pd.read_csv('TSLA_test3.csv')
# TrainData

# X_train = TrainData.iloc[:, :1]
# # X_train=X_train.values
# y_train = TrainData.iloc[:, 4:5]

# X_test = TestData.iloc[:, :1].values
# # X_train=X_train.values
# y_test = TestData.iloc[:, 4:5].values



print('X_train',X_train)
print('y_train',y_train)




```

```python
# print('X_test',X_test)
# print('y_test',y_test)
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression  
#  # create RF regressor object 
regressor = RandomForestRegressor() 
regressor.fit(X_train, y_train)

y_pred = regressor.predict(y_test)
```

```python
plt.rcParams["figure.figsize"] = (20,8)

plt.plot(X_train, y_train, color = 'red', label="Actual Train")
plt.plot(X_test,y_test, color='green', label="Actual Test")
plt.plot(X_test,y_pred, color='blue', label="Prediction")

plt.title('Tesla Stock Prices 2015-Today')
plt.xlabel('Days')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python
from sklearn import metrics
print('R-Square Error:', metrics.r2_score(y_test , y_pred))
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test , y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test , y_pred)))
print('Mean Squared Error:', metrics.mean_squared_error(y_test , y_pred))
```
